```python
import autograd.numpy as np
import pandas as pd
```

```python
r = 3
h = 10
a = 7
d = 5
ho = 24

H = np.random.randn(h, r)
A = np.random.randn(a, r)
D = np.random.randn(d, r)
T = np.random.randn(a, ho, r)
```

```python
np.einsum('Hr, Ar, Dr, Tr ->HADT', H, A, D, T[0])
```

```python
pred = np.einsum('Hr, Ar, Dr, Tr ->HADT', H, A, D, T[0])
for i in range(a):
    pred[:, i:i+1, :, :] = np.einsum('Hr, Ar, Dr, Tr ->HADT', H, A[i:i+1], D, T[i])
```

```python
np.allclose(np.einsum('Hr, Ar, Dr, ATr ->HADT', H, A, D, T), pred)
```

```python
def cost(H, A, D, T, tensor):
    pred = np.einsum('Hr, Ar, Dr, ATr ->HADT', H, A, D, T)
    
    return ((tensor-pred)**2).sum()
    
```

```python
tensor = np.random.randn(h, a, d, ho)
```

```python
from autograd import grad
```

```python
g = grad(cost, argnum=[0, 1, 2, 3])

lr = 1

dh, da, dd, dt = g(H, A, D, T, tensor)
```
