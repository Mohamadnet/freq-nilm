```python
import pickle
import numpy as np
import pandas as pd
```

```python
f = open('../code/rnn-results.txt', 'r')
lines = f.readlines()
```

```python
line_num = 1
count = 0
orders = []
while line_num < 5761:
    order = lines[line_num][:-1]
    print order
    orders.append(order)
    
    error = []
    for j in range(1, 7):
        error.append(float(lines[line_num+j].split()[1]))
    df = pd.DataFrame(error).T
    df = df.rename(index={0:count})
    count += 1
    
    if line_num == 1:
        error_df = df
    else:
        error_df = pd.concat([error_df, df])
    
    line_num += 8
    
    
```

```python
error_df = error_df.rename(columns={0:'dw', 1:'fridge', 2:'hvac', 3:'mw', 4:'oven', 5:'wm'})
```

```python
appliance = 'fridge'
error_df.nsmallest(10, appliance)
```
