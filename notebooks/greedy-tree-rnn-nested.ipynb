```python
import numpy as np
```

```python
appliance_list = ['hvac', 'fridge', 'dr', 'dw', 'mw']
```

```python
fold_num = 3
dataset = 2
```

```python
raw_error = {}
for app1 in ['hvac', 'fridge', 'mw']:
    raw_error[app1] = {}
    for app2 in [app for app in appliance_list if app != app1]:
        raw_error[app1][app2] = {}
        print (app1, app2)
        for cell_type in ['GRU', 'RNN', 'LSTM']:
            raw_error[app1][app2][cell_type] = {}
            for hidden_size  in [20, 50, 100]:
                raw_error[app1][app2][cell_type][hidden_size] = {}
                for num_layers in range(1,4):
                    if hidden_size == 100 and num_layers == 2:
                        continue
                    if hidden_size == 100  and num_layers == 3:
                        continue
                    raw_error[app1][app2][cell_type][hidden_size][num_layers] = {}
                    for bidirectional in [True]:
                        raw_error[app1][app2][cell_type][hidden_size][num_layers][bidirectional] = {}
                        for lr in [0.01, 0.1, 1.0]:
                            raw_error[app1][app2][cell_type][hidden_size][num_layers][bidirectional][lr] = {}
                            for iters in [1000, 2000, 3000]:
                                try:
                                    raw_error[app1][app2][cell_type][hidden_size][num_layers][bidirectional][lr][iters] = \
                                    np.load("../code/baseline/rnn-nested-cv/valid-error-{}-{}-{}-{}-{}-{}-{}-{}-0.0-[\'{}\', \'{}\'].npy"\
                                            .format(fold_num, dataset, cell_type, hidden_size, num_layers, bidirectional, lr, iters, app1, app2)).item()
                                except:
#                                     continue
                                    print ("python rnn-nested-cv-new.py", fold_num, dataset, cell_type, hidden_size, num_layers, bidirectional, lr, iters, 0, app1,app2)
```

```python
appliance_contri = {'hvac':0.83003428, 'fridge':0.0827564, 'dr':0.06381463, 'dw':0.01472098, 'mw':0.00867371}
```

```python
error = {}
for app1 in ['hvac', 'fridge', 'mw']:
    error[app1] = {}
    for app2 in [app for app in appliance_list if app != app1]:
        error[app1][app2] = {}
        for cell_type in ['GRU', 'RNN', 'LSTM']:
            error[app1][app2][cell_type] = {}
            for hidden_size  in [20, 50, 100]:
                error[app1][app2][cell_type][hidden_size] = {}
                for num_layers in range(1,4):
                    if hidden_size == 100 and num_layers == 2:
                        continue
                    if hidden_size == 100  and num_layers == 3:
                        continue
                    error[app1][app2][cell_type][hidden_size][num_layers] = {}
                    for bidirectional in [True]:
                        error[app1][app2][cell_type][hidden_size][num_layers][bidirectional] = {}
                        for lr in [0.01, 0.1, 1.0]:
                            error[app1][app2][cell_type][hidden_size][num_layers][bidirectional][lr] = {}
                            for iters in [1000, 2000, 3000]:
                                try:
                                    e1 = raw_error[app1][app2][cell_type][hidden_size][num_layers][bidirectional][lr][iters][app1]/appliance_contri[app1]
                                    e2 = raw_error[app1][app2][cell_type][hidden_size][num_layers][bidirectional][lr][iters][app2]/appliance_contri[app2]
                                    error[app1][app2][cell_type][hidden_size][num_layers][bidirectional][lr][iters] = e1+e2
                                    if e1+e2 < 1125:
                                        print (app1, app2, fold_num, dataset, cell_type, hidden_size, num_layers, bidirectional, lr, iters, e1+e2)
                                except:
                                    continue
```
